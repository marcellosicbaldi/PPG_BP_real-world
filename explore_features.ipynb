```python
features = pd.DataFrame()
fiducials = pd.DataFrame()
for i, sub_ID in enumerate(sub_names): 

    fiducials_sub = pd.read_csv(data_path + sub_ID + '/' + '/fiducials.csv')
    fiducials_sub['sub_ID'] = sub_ID

    features_sub = pd.read_csv(data_path + sub_ID + '/' + '/features.csv')
    features_sub['sub_ID'] = sub_ID

    features = pd.concat([features, features_sub], axis = 0)

    fiducials = pd.concat([fiducials, fiducials_sub], axis = 0)
```

Let's start from the Inflection Point Area (IPA) and the Reflection Index (RI, Adp/Asp)

- IPA is an indicator of total peripheral resistance 
- RI measures the stiffness level of small to moderate arteries

```python
IPA = features[["IPA", "sub_ID"]]
IPA["onset"] = fiducials["onset"].copy()
# IPA = IPA.set_index("sub_ID")
# IPA.drop("sub_ID", axis = 1, inplace = True)

RI = features[["Adp/Asp", "sub_ID"]]
RI["onset"] = fiducials["onset"].copy()
# RI = RI.set_index("sub_ID")
# RI.drop("sub_ID", axis = 1, inplace = True)

Asp = features[["Asp", "sub_ID"]]
Asp["onset"] = fiducials["onset"].copy()

Adp = features[["Adp", "sub_ID"]]
Adp["onset"] = fiducials["onset"].copy()

Adn = features[["Adn", "sub_ID"]]
Adn["onset"] = fiducials["onset"].copy()

SysAmp = features[["Asp", "sub_ID"]]
SysAmp["onset"] = fiducials["onset"].copy()
```

```python
def remove_outliers(feature, m = 1.5):
    """
    Remove outliers from the data (defined as data points that are 1.5 times the standard deviation away from the mean)
    Input:
    - feature: pandas DataFrame containing the feature to be cleaned
    - m: float, threshold for the outlier detection
    Output:
    - feature: pandas DataFrame, cleaned feature
    """
    feature = feature[~((feature - feature.mean()).abs() > m * feature.std())]
    return feature
```

```python
# remove outliers for each sub from SysAmp
for sub_ID in sub_names:
    SysAmp_copy = SysAmp.copy()
    SysAmp.loc[SysAmp["sub_ID"] == sub_ID, "Asp"] = remove_outliers(SysAmp_copy[SysAmp_copy["sub_ID"] == sub_ID]["Asp"])
```

```python
plt.figure(figsize=(18, 10))
sns.boxplot(data = SysAmp, x = "sub_ID", y = "Asp", showfliers = False)
plt.xlabel('Subject ID', fontsize = 20)
plt.ylabel('Systolic Amplitude', fontsize = 20)
plt.yticks(fontsize = 18)
plt.title('Systolic Amplitude for each subject', fontsize = 20)
```

```python
# Explore IPA and RI --> distribition for each subject
plt.figure(figsize=(18, 10))
sns.boxplot(data = IPA, x = "sub_ID", y = "IPA", showfliers = False)
plt.figure(figsize=(18, 10))
sns.boxplot(data = RI, x = "sub_ID", y = "Adp/Asp", showfliers = False)
plt.figure(figsize=(18, 10))
sns.boxplot(data = Asp, x = "sub_ID", y = "Asp", showfliers = False)
plt.figure(figsize=(18, 10))
sns.boxplot(data = Adp, x = "sub_ID", y = "Adp", showfliers = False)
plt.figure(figsize=(18, 10))
sns.boxplot(data = Adn, x = "sub_ID", y = "Adn", showfliers = False)
```

```python
# Plot the time course of IPA and RI for each subject
plt.figure(figsize=(18, 10))
sns.lineplot(data = IPA[IPA["sub_ID"] == "1_ID"], x = "onset", y = "IPA", hue = "sub_ID", marker = "o")
```
